## Trajectories for the 10 bouts before & after rotation
This figure is not in the article, but serves to get a quick sense of how the animal's behavior changes from before to after maze rotation.

````python
# Draw nose trajectories color coded with time: 10 bouts before and 10 bouts after rotation
from matplotlib.collections import LineCollection
from matplotlib import cm

def DrawBout(nickname,b,ax):
    tf=LoadTraj(nickname+'-tf')
    x=-0.5+15*tf.ke[b][:,0]
    y=-0.5+15*tf.ke[b][:,1]
    t=np.linspace(0,1,x.shape[0]) # time variable
    points=np.array([x,y]).transpose().reshape(-1,1,2) # array of points
    segs=np.concatenate([points[:-1],points[1:]],axis=1) # array of segments
    lc=LineCollection(segs,cmap=plt.get_cmap('viridis'),linewidths=1.5) # collect segments
    lc.set_array(t) # color the segments by the time variable
    plot(ma.wa[:,0],ma.wa[:,1],fmts=['k-'],equal=True,linewidth=1,yflip=True,
              xhide=True,yhide=True,axes=ax) # Draw the maze outline    
    re=[[-0.5,0.5,1,1],[-0.5,4.5,1,1],[-0.5,8.5,1,1],[-0.5,12.5,1,1],
       [2.5,13.5,1,1],[6.5,13.5,1,1],[10.5,13.5,1,1],
       [13.5,12.5,1,1],[13.5,8.5,1,1],[13.5,4.5,1,1],[13.5,0.5,1,1],
       [10.5,-0.5,1,1],[6.5,-0.5,1,1],[2.5,-0.5,1,1],
       [6.5,1.5,1,1],[6.5,11.5,1,1],[10.5,5.5,1,1],[10.5,7.5,1,1],
       [5.5,4.5,1,1],[5.5,8.5,1,1],[7.5,4.5,1,1],[7.5,8.5,1,1],[2.5,5.5,1,1],[2.5,7.5,1,1],
       [-0.5,2.5,3,1],[-0.5,10.5,3,1],[11.5,10.5,3,1],[11.5,2.5,3,1],[5.5,0.5,3,1],[5.5,12.5,3,1],
       [7.5,6.5,7,1]]
    for r in re:
        rect=patches.Rectangle((r[0],r[1]),r[2],r[3],linewidth=1,edgecolor='lightgray',facecolor='lightgray')
        ax.add_patch(rect)
    ax.axis('off'); # turn off the axes
    lines=ax.add_collection(lc); # add the collection to the plot
````

## Speed
### For each animal

````python
# compute speed on same time axis before & after rotation
# Names=[y for x in RotNames for y in [x+'a',x+'b']]
bw=100 # bin width in s
tm=5000 # max time before or after to be analyzed
col=['r','g','b','m']
fig,ax=plt.subplots(4,1,figsize=(8,8))
for i,nickname in enumerate(RotNames):
    ct=np.array([]) # step times before & after rotation, measured in time-in-maze
    for s in ['a','b']:
        tf=LoadTraj(nickname+s+'-tf')
        for j,b in enumerate(tf.ce):
            d=b[:] # cell for each frame
            d[1:]-=d[:-1] # change in cell
            d[0]=1 # first cell called changed
            t0=TimeInMaze(tf.fr[j,0],tf) # time in maze at start of the bout
            ct=np.append(ct,np.where(d>0)[0]/30+t0) # time in maze for steps in this bout
        if s=='a':
            ct-=TimeInMaze(tf.fr[-1,1],tf) # subtract time in maze at end of 'before' trajectory
    nb=int(2*tm/bw) # num bins
    bins=np.linspace(-tm,tm,nb+1,endpoint=True)
    sp,ed=np.histogram(ct,bins) # histogram all the steps between cells
    sp=sp/bw # average speed in each time bin
    hist(ed[:-1],ed,weights=sp,xlabel='Time in maze from rotation (s)',ylabel='Speed (cells/s)',
        xlim=[-tm,tm],ylim=[0,2.5],legend=[nickname],color=[col[i]],
        loc='upper left',axes=ax[i]); # draw speed vs time
    plot([0,0],[0,2.5],fmts=['k:'],linewidth=1,axes=ax[i]) # line for time = 0 
for i in range(3):
    ax[i].set_xlabel('')
    ax[i].set_xticklabels([])
#     ax[i].get_xaxis().set_visible(False)
plt.savefig('figs/RotationSpeedEach.pdf',bbox_inches = "tight")
plt.show()
````

### Figure 4–figure supplement 2. Average speed over animals

````python
# Combine all steps between cells and plot cumulative rate, divided by the number of animals in the maze at that point.
t0=[];t1=[] # start and end of each mouse's time in maze
for i,nickname in enumerate(RotNames):
    tfa=LoadTraj(nickname+'a-tf')
    tfb=LoadTraj(nickname+'b-tf')
    t0+=[-TimeInMaze(tfa.fr[-1,1],tfa)]
    t1+=[TimeInMaze(tfb.fr[-1,1],tfb)]
    
def MiceInMaze(ti): # returns how many animals are in the maze at time ti
    return sum([x<=ti<=y for (x,y) in zip(t0,t1)])

cs=[]
for i,nickname in enumerate(RotNames):
    ct=np.array([]) # step times before & after rotation, measured in time-in-maze
    for s in ['a','b']:
        tf=LoadTraj(nickname+s+'-tf')
        for j,b in enumerate(tf.ce):
            d=b[:] # cell for each frame
            d[1:]-=d[:-1] # change in cell
            d[0]=1 # first cell called changed
            ts=TimeInMaze(tf.fr[j,0],tf) # time in maze at start of the bout
            ct=np.append(ct,np.where(d>0)[0]/30+ts) # time in maze for steps in this bout
        if s=='a':
            ct-=TimeInMaze(tf.fr[-1,1],tf) # subtract time in maze at end of 'before' trajectory
    cs+=[ct]
xs=np.sort(np.concatenate(cs))   
ys=np.array([1/MiceInMaze(x) for x in xs]) # average number of steps
zs=np.cumsum(ys)
plot(xs,zs,
     figsize=(6,4),
    xlabel='Time in maze from rotation (s)',
    ylabel='Average steps between cells',
);
plt.show()
tbin=100 # bin size for speed measurement
tlim=5000 # time before & after rotation
ax,_,_,_=hist(xs,
     figsize=(6,3),
     weights=ys/tbin,
     bins=np.arange(-tlim,tlim+tbin,tbin),
     xlim=[-tlim,tlim],
     ylim=[0,2.2],
     xlabel='Time in maze from rotation (s)',
     ylabel='Average speed (cells/s)',
);
plot([0,0],[0,2.5],fmts=['k:'],linewidth=1,axes=ax)
plt.savefig('figs/RotationSpeedAvg.pdf',bbox_inches = "tight")
````

- The cumulative number of steps shows no kink at t=0
- The average speed (derivative of the steps) fluctuates, but nothing significant happens at t=0

## Node visits and rewards 

````python
# Find time in maze for paths to a set of nodes
def GetPathTimes(tf,lmin,nodes):
    '''
    For every node in nodes, returns a list of times of direct paths to that node with length >= lmin 
    '''
    ma=NewMaze(6)
    times=[]
    for n in nodes:
        ptn=FindPathsToNode(n,tf,ma)
        if len(ptn)>0:
            ptn=ptn[np.where(ptn[:,2]>=lmin)] # impose minimal length
            ti=[TimeInMaze(fr,tf) for fr in ptn[:,3]] # time in maze for start of path 
            times+=[ti] # add to list of lists
        else:
            times+=[[]]
    return times    
````

# Sudden insight

Here we decided to fit the event rate for rewards and long paths to water using a different model with a sigmoid dependence on time. 

\begin{equation}
r\left( t \right) = {r_{\text{i}}} + \frac{{{r_{\text{f}}} - {r_{\text{i}}}}}{2}\operatorname{erf} \left( {\frac{{t - {t_{\text{s}}}}}{w}} \right)
\end{equation}

where

$$ \operatorname{erf} \left( x \right) = \frac{2}{{\sqrt \pi  }}\int\limits_0^x {{e^{ - {x^2}}}{\text{d}}x} $$

The rate begins at a low initial level $r_{\text{i}}$, reflecting chance occurrence of the event, and saturates at a high final level $r_{\text{f}}$, limited for example by the animal's walking speed. The other two parameters are the time $t_{\text{s}}$ of half-maximal rate change, and the width $w$ over which that rate change takes place. A sudden change in the event rate would correspond to $w=0$.



## Applying a sigmoid model to the event rate

````python
import scipy.special as sc

def rate_erf(a,b,c,d,t): # rate function
    return a*sc.erf(c*(t-b))+a+d # both a and d positive

def integ_erf(a,b,c,d,t): # integral of the rate function
    z=c*(t-b)
    return a/c*(z*sc.erf(z)+np.exp(-z*z)/np.sqrt(np.pi))+(a+d)*t

def LogLik_erf(x,t,Tl,Th): # negative log likelihood
    '''
    Negative of the Log likelihood of sigmoid rate model
    x = parameters (a,b,c,d). Note c=1.0 is not allowed.
    t = array of event times
    Tl,Th = low & high ends of the time range
    '''
    a,b,c,d=x
    return integ_erf(a,b,c,d,Th)-integ_erf(a,b,c,d,Tl)-np.sum(np.log(rate_erf(a,b,c,d,t)))

def EstimateErf(ti,t0,t1):
    '''
    fit the events ti[] in [t0,t1] with a error function rate function rate_erf
    maximize the likelihood function
    return parameters of the sigmoid
    '''
    from scipy.optimize import minimize
    
    t=ti[np.where(np.logical_and(ti>=t0,ti<=t1))] # restrict to [t0,t1]
    u=(t-t0)/(t1-t0) # scale data into [0,1] so parameters are more comparable, makes fits converge
    x0=[len(t)/2,0.5,2,len(t)/2] # initialize parameters
    res = minimize(fun=LogLik_erf,x0=x0,args=(u,0,1),
                   method='SLSQP',tol=1e-8,
                   bounds=[(0,None),(0.0,None),(0.01,500),(0,None)]) 
    a=res.x[0]/(t1-t0) # scale results back into [0,T[
    b=res.x[1]*(t1-t0)+t0
    c=res.x[2]/(t1-t0) 
    d=res.x[3]/(t1-t0) 
    return (a,b,c,d,-LogLik_erf((a,b,c,d),t,t0,t1))
````

````python
# Fit the sigmoid model parameters to rewards and long water paths for all mice
lmin=7 # minimal length of water path
t0=0 # time in maze at start of fitting interval
t1=5000 # time in maze at left end of fitting interval
rm=[] # list of parameters for the rewards model
wm=[] # list of parameters for the water path model
for nickname in RewNames:
    tf=LoadTraj(nickname+'-tf')
    re = np.array([y[0]+tf.fr[i,0] for i,r in enumerate(tf.re) for y in r])
    ti=np.array([TimeInMaze(r,tf) for r in re]) # times of rewards
    a,b,c,d,_=EstimateErf(ti,t0,t1) # fit with sigmoid
    rm+=[(a,b,c,d)] # add to list
    ptn_1 = FindPathsToNode(116,tf,ma) # water paths
    ptn_m = ptn_1[np.where(ptn_1[:,2]>=lmin)] # impose minimal length
    ti=np.array([TimeInMaze(p,tf) for p in ptn_m[:,3]]) # times of those paths
    a,b,c,d,_=EstimateErf(ti,t0,t1) # fit with sigmoid
    wm+=[(a,b,c,d)] # add to list
with open('outdata/RewardsWaterPathsErfModel','wb') as f: # save to file
    pickle.dump((RewNames,rm,wm),f) 
````
