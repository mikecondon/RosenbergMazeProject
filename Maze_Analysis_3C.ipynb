# Modes of behavior
Divide each bout of the trajectory into
- Leave: direct path to the exit at the end of the bout
- Drink: all direct paths to water
- Explore: everything else

A direct path to a target is the longest sequence of nodes without reversals that ends at the target.

Goals: 
- determine fraction of time/nodes in each mode
- determine transition rates (per time/node) between modes
- restrict analysis to one mode, e.g. concatenate all the paths in exploration mode and analyze their efficiency

Method:
- Call each piece of a bout that corresponds to one mode a 'clip' (as in cutting a video recording into clips)
- For each bout identify all the clips and make an array of [mode, start node index, number of nodes]. Combine those arrays into a list. 

Algorithm:
- Each clip should include start and end node. The end node of one is the start node of the next.
- Start with the exit. FInd the direct path. That defines the 'leave mode' clip for this bout. Find the start node of that clip.
- If the start node is the water port, find the direct path that leads there: 'drink' clip
- If it's some other node, trace back to either the exit or the water port: 'explore' clip
- If it's the exit, move on to the next bout.
- Sort all the clips in chronological order.

The routine `SplitModeClips(tf,ma,re=True)` takes care of all that.

## Mode analysis for all animals
For each animal we want:
- Number of nodes and frames in each mode
- Transition probabilities between modes
- Efficiency analysis (New nodes vs all nodes) for exploration mode specifically

```python
# Define the 3 states
leave=0; drink=1; explore=2 # the 3 modes
```

```python
# Mode analysis for all animals
Names=RewNames+UnrewNamesSub
k=len(RewNames)
for i,nickname in enumerate(Names):
    tf=LoadTraj(nickname+'-tf')
    cl=SplitModeClips(tf,ma,re=i<k) # find the clips; no drink mode for unrewarded animals
    ti=np.array([tf.no[c[0]][c[1]+c[2],1]-tf.no[c[0]][c[1],1] for c in cl]) # duration in frames of each clip
    nn=np.array([np.sum(cl[np.where(cl[:,3]==leave)][:,2]),
                 np.sum(cl[np.where(cl[:,3]==drink)][:,2]),
                 np.sum(cl[np.where(cl[:,3]==explore)][:,2])]) # number of node steps in each mode
    nf=np.array([np.sum(ti[np.where(cl[:,3]==leave)]),
                 np.sum(ti[np.where(cl[:,3]==drink)]),
                 np.sum(ti[np.where(cl[:,3]==explore)])]) # number of frames in each mode
    tr=np.zeros((3,3)) # number of transitions between the 3 modes
    for i in range(1,len(cl)):
        tr[cl[i-1,3],cl[i,3]]+=1  
    ce=cl[np.where(cl[:,3]==explore)] # clips of exploration
    ne=np.concatenate([tf.no[c[0]][c[1]:c[1]+c[2],0] for c in ce]) # nodes excluding the last state in each clip
    le=6 # end nodes only
    ln=list(range(2**le-1,2**(le+1)-1)) # list of node numbers in level le
    ns=ne[np.isin(ne,ln)] # restricted to desired nodes
    wcn=NewNodes4(ns,nf[2]/len(ns)) # compute new nodes vs all nodes for exploration mode only
    with open('outdata/'+nickname+'-Modes1', 'wb') as f: # save in one file per animal
        pickle.dump((nn,nf,tr,wcn),f)
```

### Mode fraction vs time

```python
# Histogram time in each mode for rewarded animals
# Each clip records the bout number, start state in the bout, number of steps, and mode
Names=RewNames
te=25000 # max time
nb=50 # number of bins
tl=np.linspace(0,te,nb,endpoint=False) # left end of each bin
tb=te/nb # bin width
ti=np.zeros((len(Names),nb,3))
for i,nickname in enumerate(Names):
    tf=LoadTraj(nickname+'-tf')
    cl=SplitModeClips(tf,ma,re=True) # find the clips
    for c in cl:
        b=c[0] # bout
        n0=tf.no[b][c[1]] # start node
        b0=(tf.fr[b,0]+n0[1])/30 # time of clip start in s
        n1=tf.no[b][c[1]+c[2]] # end node
        b1=(tf.fr[b,0]+n1[1])/30 # time of clip end in s
        m=c[3] # mode
        if b0<te: # if left end is before the max time
            j0=int(b0//tb) # bin containing left end
            if b1>te: # if right end is after the max time
                b1=te-1e-8 # set it to the max time minus a little for comfort
            j1=int(b1//tb) # bin containing right end
            if j1>j0: # if clip ends in a different bin
                ti[i,j0,m]+=(j0+1)*tb-b0 # part of the first bin
                ti[i,j0+1:j1,m]+=tb # all of the intervening bins
                ti[i,j1,m]+=b1-j1*tb # part of the last bin
            else: # if clip starts and ends in same bin
                ti[i,j0,m]+=b1-b0
ti/=tb # normalize to bin width to get fraction of time in each bin
ts=np.sum(ti,axis=-1) # sum the 3 time fractions in each bin
tf=ti/(ts.reshape(ts.shape+(1,))+1e-8) # fraction in each mode in each bin (avoid divide by zero)
```

## Fit new nodes curve with 2-parameter function

We will fit the curve of new nodes vs all nodes with this 2-parameter function:

$$y= 64 \left( 1-\frac{1}{1+\frac{z+b z^2}{1+b}} \right)$$

$$z= x \left/ a \right.$$

For small $x$ this grows ~$x$ as required. It equals 1/2 at $x=a$. And for large $x$ it can grow faster.

The relative uncertainty in the two parameter $a$ and $b$ can be measured by dividing the area of the error ellipse by the product of the 2 means.

The area of the error ellipse is the determinant of the covariance matrix, `np.linalg.det(pcov)`.

The product of the means is `np.prod(popt)`.

Then take the square root of that to get the relative uncertainty.

```python
# efficiency fits for all animals, keeping track of fit error
def func(x, a, b): # the 2-parameter fit function
    z=x/a
    return 64*(1-1/(1+(z+b*z**3)/(1+b)))

a=70 # starting value
b=0.1 # starting value
Names=RewNames+UnrewNamesSub
pars=np.zeros((len(Names),2))
errs=np.zeros((len(Names),2,2))
for i,nickname in enumerate(Names):
    with open('outdata/'+nickname+'-Modes1', 'rb') as f: # includes the precomputed new nodes cuve
        nn1,nf1,tr1,wcn1=pickle.load(f)
    xdata=wcn1[1]
    ydata=wcn1[2] 
    popt, pcov = curve_fit(func, xdata, ydata, p0=[a,b])
    pars[i]=popt
    errs[i]=pcov
with open('outdata/'+'ExplPars3', 'wb') as f: # save fit parameters and their errors, one file
    pickle.dump((Names,pars,errs),f)
```

## Measure the 4 turning bias parameters during exploration

This section focuses on the decision biases of the animal during exploration mode.

The function `ComputeFourBiasClips2(tr,ma,clips,mode=2)` takes care of that.

```python
# Derives 4 bias parameters from an entire mouse trajectory for all nodes except entry and end nodes
def ComputeFourBias2(tf,ma):
    '''
    Derives 4 bias parameters from mouse trajectory for all nodes except entry and end nodes
    This version spots occasions where a bias can't be estimated.
    Bf = prob of going forward when arriving from bottom of the T
    Ba = prob of making an alternating (instead of repeating) turn when going forward from bottom
    Lf = prob of going forward when arriving from L branch (or R branch)
    Lo = prob of turning out (instead of in) when going forward from L (or R)
    '''
    sta,trb=SecondTransProb(tf,ma)
    bi=np.array([Bias2(i,ma,trb,alt=True) for i in range(2**ma.le-1)]) # compute 6-parameter bias for each node
    va=np.sum(bi>-1,axis=0) # number of valid bias values in each category
    Bf=-1;Ba=-1;Lf=-1;Lo=-1 # return -1 if no valid measurement
    if va[0,0]>0:
        Bf=np.average(bi[:,0,0],weights=(bi[:,0,0]>-1))
    if va[0,1]>0:
        Ba=np.average(bi[:,0,1],weights=(bi[:,0,1]>-1))
    if va[1,0]>0 and va[2,0]>0:
        Lf=0.5*(np.average(bi[:,1,0],weights=(bi[:,1,0]>-1))+np.average(bi[:,2,0],weights=(bi[:,2,0]>-1)))
    if va[1,1]>0 and va[2,1]>0:
        Lo=0.5*(np.average(bi[:,1,1],weights=(bi[:,1,1]>-1))+np.average(bi[:,2,1],weights=(bi[:,2,1]>-1)))
    return [Bf,Ba,Lf,Lo]
```

```python
# Derive 4 bias parameters from mouse trajectory, restricted to clips of type mode
# This version also returns the standard error across nodes
def ComputeFourBiasClips2(tr,ma,clips,mode=2):
    '''
    Computes 4 bias parameters from trajectory tr in maze ma
    Limits itself to the clips of behavior type mode    
    Bf = prob of going forward when arriving from bottom of the T
    Ba = prob of making an alternating (instead of repeating) turn when going forward from bottom
    Lf = prob of going forward when arriving from L branch (or R branch)
    Lo = prob of turning out (instead of in) when going forward from L (or R)
    '''
    sta,trb=SecondTransProbClips(tr,ma,clips,mode)
    # compute 6-parameter bias for each node without entry or end nodes
    bi=np.array([Bias2(i,ma,trb,alt=True) for i in range(1,2**ma.le-1)]) 
    Bf=-1;Ba=-1;Lf=-1;Lo=-1 # return -1 if no valid measurement
    Bfs=-1;Bas=-1;Lfs=-1;Los=-1 # return -1 if no valid measurement
    x=bi[:,0,0]
    i=x>-1
    if len(x[i])>0:
        Bf=np.mean(x[i])
        Bfs=np.std(x[i])
    x=bi[:,0,1]
    i=x>-1
    if len(x[i])>0:
        Ba=np.mean(x[i])
        Bas=np.std(x[i])
    x=np.concatenate((bi[:,1,0],bi[:,2,0]))
    i=x>-1
    if len(x[i])>0:
        Lf=np.mean(x[i])
        Lfs=np.std(x[i])             
    x=np.concatenate((bi[:,1,1],bi[:,2,1]))
    i=x>-1
    if len(x[i])>0:
        Lo=np.mean(x[i])
        Los=np.std(x[i])             
    return [Bf,Ba,Lf,Lo],[Bfs,Bas,Lfs,Los]
```

```python
# Compute 2nd order transition probabilities within clips of one behavior mode
def SecondTransProbClips(tr,ma,clips,mode=2):
    '''
    Computes 2nd order transition probabilities among nodes of maze ma in trajectory tr
    Limits itself to the clips of behavior type mode
    Returns arrays sta(n,3) and trb(n,3,3) with n=# of nodes
    sta[i,:] the 3 nodes connected to node i in order o,L,R; i.e. parent, left child, right child
    trb[i,j,k] is the prob of transition from state i to sta[i,k] given prior state was sta[i,j]
    '''
    sta=TransMatrix(ma) # array of nodes connected to each node, in order parent, left child, right child
    # make array of transition probabilities based on current and preceding state 
    ta = TallyStringsClips(tr,clips,mode,3) # all occurrences of strings up to length 3
    trb=np.zeros((len(ma.ru),3,3)) # 3D array containing transition probability depending on 2-string
    for i in range(len(ma.ru)): # i is current state
        for j,sj in enumerate(sta[i]): # sta[i,j] is preceding state
            for k,sk in enumerate(sta[i]): # sta[i,k] is next state
                if (sj,i,sk) in ta[2]:
                    trb[i,j,k]=ta[2][(sj,i,sk)]/ta[1][(sj,i)]
    return sta,trb
```

```python
# Tally occurrences for each j-string within clips of one behavior mode
def TallyStringsClips(tr,clips,mode=2,m=5):
    '''
    Produces m dictionaries that give the number of occurrences for each j-string 
    up to j=m in the trajectory tr.
    The strings of different length are aligned to all share the first element.
    Limits itself to the clips of behavior type mode
    Returns: a list of dictionaries of type s[string tuple]=number.
    '''
    se=[{} for i in range(m)] # se[j] tallies (j+1)-strings
    nt=tr.no
    for cl in clips:
        if cl[3]==mode:
            b=nt[cl[0]][cl[1]:cl[1]+cl[2]+1] # the piece of bout of this clip, incl first and last node
            for i in range(0,len(b)-m+1): # first position of the string
                for j in range(m): # j+1 = length of string
                    s=tuple(b[i:i+j+1,0]) # j+1-string starting at i
                    if s in se[j]: # tally into the j-th dictionary
                        se[j][s]+=1
                    else:
                        se[j][s]=1
    return se
```

```python
# A function to compute 6-parameter bias
def Bias2(i,m,trb,alt=False):
    '''
    Computes 6 biases for node i of maze m based on transition probs in trb
    Can enter the T junction from bottom (B), left (L), or right (R)
    From each direction, can either go forward (f) or reverse back to the preceding state.
    If go forward from L or R, can go out (o) or into the maze
    If go forward from B, can go left (l) or right.
    Alternatively, if alt==True score this as an alternating (a) vs same direction turn.
    So the components of the bias are: Bf, Bl, Lf, Lo, Rf, Ro
    If a particular entry (B, L, or R) never occurs then returns -1 for those 2 bias values.
    '''
    def Norm(x):
        if x[0]==0:
            return 0
        else:
            return x[0]/sum(x)
    tr=trb[i] # 2-state transitions for this node

    B=np.sum(tr[0])
    if B>0:
        Bf=(tr[0,1]+tr[0,2])/B # Bf = forward bias from B
        if Bf>0:
            if alt and StepType(m.pa[i],i,m)==0: # an L node, so 'alt' refers to right turns
                Bl=tr[0,2]/(tr[0,2]+tr[0,1])
            else:
                Bl=tr[0,1]/(tr[0,2]+tr[0,1]) # Bl = left bias when stepping forward from B
        else:
            Bl=-1 # can't evaluate Bl bias
    else:
        Bf=-1;Bl=-1 # can't evaluate Bf and Bl bias

    L=np.sum(tr[1])
    if L>0:
        Lf=(tr[1,0]+tr[1,2])/L # Lf = forward bias from L
        if Lf>0:
            Lo=tr[1,0]/(tr[1,0]+tr[1,2]) # Lo = outward bias if forward from L
        else:
            Lo=-1 # can't evaluate Lo bias
    else:
        Lf=-1;Lo=-1 # can't evaluate Lf and Lo bias
            
    R=np.sum(tr[2])
    if R>0:
        Rf=(tr[2,0]+tr[2,1])/R # Rf = forward bias from R
        if Rf>0:
            Ro=tr[2,0]/(tr[2,0]+tr[2,1]) # Ro = outward bias if forward from R
        else:
            Ro=-1 # can't evaluate Ro bias
    else:
        Rf=-1;Ro=-1 # can't evaluate Rf and LRo bias
    return np.array([[Bf, Bl],[Lf, Lo],[Rf, Ro]])

```

## Models for sudden insight

We are looking at events whose rate seems to suddenly step up at a specific moment in time. Here we try to model that event rate 3 different ways:

1. A sudden step: The rate is constant before the step and changes to another constant value after the step  (3 parameters)

2. A gradual ramp: The rate increases gradually as a quadratic function of time (3 parameters)

3. A sigmoid step: The rate changes from an early value to a late value with a sigmoid step (4 parameters)

The data are a set of $n$ event times $t_i$ in the obervation interval $[0,T]$. We model the event train as a Poisson point process with instantaneous rate $r(t)$. The likelihood of the data given the rate function is 

$$ L\left[ {r\left( t \right)} \right] = {e^{ - \int\limits_0^T {r\left( t \right)dt} }}\prod\limits_i {r\left( {{t_i}} \right)} $$

and the log likelihood is 

$$ \ln L = \sum\limits_i^{} {\ln } r\left( {{t_i}} \right) - \int\limits_0^T {r\left( t \right)dt} $$

For each of the models we maximize $\ln L$ over the parameters. Then we compare the likelihoods for different models (step vs ramp). And we compare the likelihood for different parameters, e.g. to get their uncertainty.

## Step rate model

Rate model:

$$ r\left( t \right) = \left\{ {\begin{array}{*{20}{c}}
  {a,t < c} \\ 
  {b,t \geqslant c} 
\end{array}} \right. $$

$$ \int\limits_0^T {r\left( t \right)dt}  = ac + b\left( {T - c} \right) $$

Log likelihood:

$$ \ln L = \sum\limits_{{t_i} < c}^{} {\ln a}  + \sum\limits_{{t_i} > c}^{} {\ln b}  - ac - b\left( {T - c} \right) $$

For a given $c$, the optimal values of $a$ and $b$ are found by differentiating $\ln L$:

$$ {a^*} = \frac{{{n_ < }\left( c \right)}}{c},{b^*} = \frac{{{n_ > }\left( c \right)}}{{T - c}}$$

where ${n_ < }\left( c \right)$ and ${n_ > }\left( c \right)$ are respectively the numbers of events before and after the step time $c$,

$${n_ < }\left( c \right) = \sum\limits_{{t_i} < c} 1 ,{n_ > }\left( c \right) = \sum\limits_{{t_i} > c} 1 $$

So the likelihood can be evaluated as a function of $c$ only:

$$ \ln L\left( c \right) = {n_ < }\left( c \right)\ln {n_ < }\left( c \right) + {n_ > }\left( c \right)\ln {n_ > }\left( c \right) - {n_ < }\left( c \right)\ln c - {n_ > }\left( c \right)\ln \left( {T - c} \right) - n $$

where $n$ is the total number of events.


`EstimateStep(ti,t0,t1)` evaluates the likelihood function, returns the ML values of the parameters $a$, $b$, and $c$, along with the mean and standard deviation of $c$ over the likelihood function.

```python
def EstimateStep(ti,t0,t1):
    '''
    ti = array of event times
    analysis restricted to [t0,t1]
    approximate the frequency of events as two rates rA and rB with a step at t
    compute likelihood function for t
    returns:
    tML,tMe,tSD: mode, mean, and SD of t over the likelihood function
    rA,rB: the two rates at the peak of the likelihood function
    ML: log likelihood at the peak of the likelihood function
    '''
    dt=(t1-t0)/1000 # time resolution on the step time
    hi,bi=np.histogram(ti,bins=np.linspace(t0,t1,(t1-t0)/dt+1)) # histogram the events in [t0,t1]
    cu=np.cumsum(hi) # cumulative
    n=cu[-1] # total number of events
    i=cu[:-1] # number of events before time t
    t=bi[1:-1] # left edges of all the bins, excepting t0 and t1
    Li=xlogx(i)-i*Ln(t-t0)+xlogx(n-i)-(n-i)*Ln(t1-t)-n # log likelihood
    jML=np.argmax(Li) # index of highest log likelihood
    ML=Li[jML] # log likelihood at the peak
    tML=t[jML] # ML value for the step time t
    rA=i[jML]/(tML-t0) # ML value for rA
    rB=(n-i[jML])/(t1-tML) # ML value for rB
    Pr=np.exp(Li-np.max(Li)) # exponentiate log likelihood
    Pr/=np.sum(Pr) # normalize to a probability
    tMe=np.sum(t*Pr) # mean of the step time t
    tSD=np.sqrt(np.sum(t**2*Pr)-tMe**2) # SD of the step time t
    return tML,rA,rB,tMe,tSD,ML
```

```python
# test step model fit on C1
tf=LoadTraj('C1-tf')
ptn_1 = FindPathsToNode(116,tf,ma) # water
lmin=7 # minimal length
ptn_m = ptn_1[np.where(ptn_1[:,2]>=lmin)] # impose minimal length
ti=np.array([TimeInMaze(p,tf) for p in ptn_m[:,3]])
t0=0
t1=5000
tML,rA,rB,tMe,tSD,MLs=EstimateStep(ti,t0,t1)
print('ML step time   tML = {:.2f}'.format(tML))
print('Rate before     rA = {:.2e}'.format(rA))
print('Rate after      rB = {:.2e}'.format(rB))    
print('Mean Step time tMe = {:.2f} ± {:.2f}'.format(tMe,tSD))
print('Likelihood         = {:.2f}'.format(MLs))
t=ti[np.where(np.logical_and(ti>=t0,ti<=t1))] # restrict to [t0,t1]
LL=np.sum(np.where(t<tML,Ln(rA),Ln(rB)))-(rA*(tML-t0)+rB*(t1-tML))
print('LL (double check)  = {:.2f}'.format(LL))
x=t
d=np.arange(len(x))+1
ys=np.where(x<tML,rA*x,rA*tML+rB*(x-tML)) # cumulative rate under step model
plot(x,[d,ys],fmts=['r-','g-'],legend=['data','step']);
```

## Quadratic rate model 

Here the rate of events increases gradually as a quadratic function of time. This has the same number of parameters as the step model.

Rate:

$$r(t)=a+bt+ct^2$$

Likelihood: 

$$L(a,b,c) = e^{{-\int_{0}^{T} r(t) dt}} \prod_{i=1}^n r(t_i)$$

\begin{aligned}
\ln{L(a,b,c)} &= -\int_{0}^{T} {r(t) dt} + \sum_{i=1}^n \ln{r(t_i)} \\
&= -\int_{0}^{T} {\left( a+bt+ct^2 \right) dt} + \sum_{i=1}^n \ln{\left( a + b t_i+c {t_i}^2 \right)}\\
&= -\left( a T +b T^2/2 + c T^3/3 \right) + \sum_{i=1}^n \ln{\left( a + b t_i+c {t_i}^2 \right)}
\end{aligned}

Maximize:

\begin{aligned}
  \frac{{\partial \ln L}}{{\partial a}} &=  - T + \sum\limits_{i = 1}^n {\frac{1}{{a + b{t_i} + ct_i^2}}}  = 0 \\
  \frac{{\partial \ln L}}{{\partial b}} &=  - \frac{{{T^2}}}{2} + \sum\limits_{i = 1}^n {\frac{{{t_i}}}{{a + b{t_i} + ct_i^2}}}  = 0 \\
  \frac{{\partial \ln L}}{{\partial c}} &=  - \frac{{{T^3}}}{3} + \sum\limits_{i = 1}^n {\frac{{t_i^2}}{{a + b{t_i} + ct_i^2}}}  = 0  
\end{aligned} 

Note that

$$a\frac{{\partial \ln L}}{{\partial a}} + b\frac{{\partial \ln L}}{{\partial b}} + c\frac{{\partial \ln L}}{{\partial c}} =  - \left( {aT + \frac{{b{T^2}}}{2} + \frac{{c{T^3}}}{3}} \right) + n = 0$$

So this gives $c$ in terms of $a$ and $b$ which leaves us with 2 equations for $a$ and $b$. 

But not clear that they have an analytical solution. 

Instead, let's brute-force this with a numerical search...

### Numerical solution

I played around considerably with the various versions of `minimize()` and settled on `method='SLSQP'` which gave robust convergence from a range of starting values.

```python
def EstimateQuad(ti,t0,t1):
    '''
    fit the events ti[] in [t0,t1] with a quadratic rate function r(t)=a+b*t+c*t**2
    maximize the likelihood function for a,b,c
    return a,b,c
    '''
    from scipy.optimize import minimize
    
    def LogLik(x,t,t0,t1):
        '''
        Negative of the Log likelihood of quadratic rate model
        '''
        a,b,c=x
        return (a*(t1-t0)+b/2*(t1**2-t0**2)+c/3*(t1**3-t0**3))-np.sum(np.log(a+b*t+c*t**2))
    
    t=ti[np.where(np.logical_and(ti>=t0,ti<=t1))] # restrict to [t0,t1]
    u=(t-t0)/(t1-t0) # scale data into [0,1] so parameters are more comparable, makes fits converge
#     x0=[len(t)/3,len(t)*2/3,len(t)*3/3] # initialize, equally across a,b,c
    x0=[len(t),0,0] # initialize, only constant term 
    res = minimize(fun=LogLik,x0=x0,args=(u,0,1),
                   method='SLSQP',tol=1e-8,
                   bounds=[(0,None),(0,None),(0,None)]) 
    c=res.x[2]/(t1-t0)**3 # scale results back into [t0,t1]
    b=res.x[1]/(t1-t0)**2-2*c*t0
    a=res.x[0]/(t1-t0)-b*t0-c*t0**2
    return a,b,c,-LogLik([a,b,c],t,t0,t1)
```

```python
# test quadratic model fit on C1
tf=LoadTraj('C1-tf')
ptn_1 = FindPathsToNode(116,tf,ma) # water
lmin=7 # minimal length
ptn_m = ptn_1[np.where(ptn_1[:,2]>=lmin)] # impose minimal length
ti=np.array([TimeInMaze(p,tf) for p in ptn_m[:,3]])
t0=0
t1=5000
a,b,c,MLq=EstimateQuad(ti,t0,t1)
print('a,b,c = {:.2e},{:.2e},{:.2e}'.format(a,b,c))
print('Likelihood         = {:.2f}'.format(MLq))
t=ti[np.where(np.logical_and(ti>=t0,ti<=t1))] # restrict to [t0,t1]
LL=np.sum(np.log(a+b*t+c*t**2))-(a*(t1-t0)+b/2*(t1**2-t0**2)+c/3*(t1**3-t0**3))
print('LL (double check)  = {:.2f}'.format(LL))
x=t
d=np.arange(len(x))+1
yq=a*(x-t0)+b/2*(x**2-t0**2)+c/3*(x**3-t0**3) # cumulative rate under quad model
plot(x,[d,yq],fmts=['r-','g-'],legend=['data','quad']);
```
